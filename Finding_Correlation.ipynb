##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
# Use only valid salary values
salary = df['ConvertedCompYearly'].dropna()

# Remove extreme outliers for readability
salary = salary[salary < salary.quantile(0.99)]

plt.figure(figsize=(10,5))
plt.hist(salary, bins=40)
plt.title("Distribution of Yearly Compensation (ConvertedCompYearly)")
plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Number of Respondents")
plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
# Filter full-time employees with valid salary
full_time = df[
    (df['Employment'] == 'Employed, full-time') &
    (df['ConvertedCompYearly'].notna())
]

median_salary = full_time['ConvertedCompYearly'].median()
median_salary
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Keep only rows with salary + country
country_pay = df[['Country', 'ConvertedCompYearly']].dropna()

# Keep top 10 countries by number of respondents
top_countries = country_pay['Country'].value_counts().head(10).index
country_pay = country_pay[country_pay['Country'].isin(top_countries)]

# Remove extreme outliers (99th percentile)
country_pay = country_pay[country_pay['ConvertedCompYearly'] < country_pay['ConvertedCompYearly'].quantile(0.99)]

plt.figure(figsize=(14,6))
country_pay.boxplot(by='Country', column='ConvertedCompYearly', rot=45)
plt.title("Compensation Distribution by Country (Top 10 Respondent Countries)")
plt.suptitle("")
plt.xlabel("Country")
plt.ylabel("Annual Compensation (USD)")
plt.tight_layout()
plt.show()
```

```python
country_pay['CountryClean'] = country_pay['Country'].replace({
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK'
})
```

```python
plt.figure(figsize=(14,6))

country_pay.boxplot(
    by='CountryClean',
    column='ConvertedCompYearly',
    rot=30,
    grid=False
)

plt.title("Compensation Distribution by Country (Top 10 Respondent Countries)")
plt.suptitle("")
plt.xlabel("Country")
plt.ylabel("Annual Compensation (USD)")
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
clean_df = df[['ConvertedCompYearly','WorkExp','JobSatPoints_1']].dropna()

# Remove extreme salary outliers
upper = clean_df['ConvertedCompYearly'].quantile(0.99)
clean_df = clean_df[clean_df['ConvertedCompYearly'] < upper]

clean_df.head()
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
# Calculate correlation matrix
corr_matrix = clean_df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].corr()

corr_matrix
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title("Correlation Heatmap of Salary, Work Experience, and Job Satisfaction")
plt.tight_layout()
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
plt.figure(figsize=(6,4))

plt.scatter(clean_df['WorkExp'], clean_df['ConvertedCompYearly'], alpha=0.5)

plt.title("Salary vs Work Experience")
plt.xlabel("Years of Work Experience")
plt.ylabel("Annual Compensation (USD)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(clean_df['JobSatPoints_1'], clean_df['ConvertedCompYearly'], alpha=0.5)

plt.title("Salary vs Job Satisfaction")
plt.xlabel("Job Satisfaction Score")
plt.ylabel("Annual Compensation (USD)")
plt.tight_layout()
plt.show()
```
